```python
import matplotlib.pyplot as plt
x = [5, 6, 7, 6, 5]
y = [0.076, 0.075, 0.065, 0.085, 0.075]
plt.plot(x, y, label='Качество вина')
plt.xlabel('Качество')
plt.ylabel('Хлориды')
plt.grid()
plt.legend()
plt.show()
```
